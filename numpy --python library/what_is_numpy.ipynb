```python
#using the python list
import numpy as np #numpy library of the python 
import time
size = 1_000_000

start = time.time()
l1 = list(range(size))
l2 = list(range(size))
start = time.time()
add = (x+y for x,y in zip(l1,l2))
end = time.time()
print(end - start)
```

```python
#using the numpy arrays 
import numpy as np
import time 
size = 1_000_000


start = time.time()
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))
start = time.time()
add = l1+l2
end = time.time()
print(end - start)
```
