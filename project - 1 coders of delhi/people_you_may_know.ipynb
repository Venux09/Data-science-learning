```python
import json


def load_data(filename):#functions for loading the data from json file
    
    with open(filename,'r') as f:
        return json.load(f)
        
def find_people_you_may_know(user_id,data):#functions for the  friends you may know 
    user_friends = {}
    for user in data["users"]:
         user_friends[user['id']] = set(user['friends'])
    if user_id not in user_friends:
        return[]

    direct_friends = user_friends[user_id]
    suggestions = {}
    for friend in direct_friends:#it means it shows the direct friends of the user_id
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends:
                #counting the individual 
                suggestions[mutual] = suggestions.get(mutual,0)+1
    sorted_suggestions = sorted(suggestions.items(), key = lambda x:x[1],reverse =True)
    return[user_id for user_id , mutual_count in sorted_suggestions]

#loading the data 
data = load_data('data.json')
user_id =1
recc = find_people_you_may_know(user_id,data)
print("You may know :",recc)


    
```
