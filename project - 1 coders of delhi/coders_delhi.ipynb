```python
import json
```

```python
def load_data(filename):
    with  open(filename ,"r") as file:
        data = json.load(file)
    return data 
```

```python
def display_user(data):
    print("User and their connections--\n")
    for user in data['users']:
        print(f"ID[{user['id']}].{user['name']} is friend with:{user['friends']} their liked pages are {user['liked_pages']}")
    print("\nthe information about the pages:-")
    for page in data['pages']:
        print(f"{page['id']} : {page['name']}")
    print("this represent the info about the pages user is liking ")
display_user(data)
```
