```python
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)


#function to find the common likiness of the page based on the
        
        #common interest of both users
def find_pages_you_might_like(user_id,data):
    #dictonary which contain  user interating with the pages 
    user_pages = {}
    for user in data['users']:
        user_pages[user['id']] = set(user['liked_pages'])
    #if user is not found return the empty list
    if user_id not in user_pages:
        return []

    #pages which are directly liked by the user 
    user_liked_pages = user_pages[user_id]#example user_liked_pages = 1 ,2  , 3 
    page_suggestion = {}

    for other_user , pages in user_pages.items():
        if other_user != user_pages:
            shared_pages =  user_liked_pages.intersection(pages)
        for page in pages:
            if pages not in user_liked_pages:
                page_suggestion[page] = page_suggestion.get(page, 0) + len(shared_pages)

    sorted_pages = sorted(page_suggestion.items(),key = lambda x:x[1],reverse = True)
    return[page_id for page_id, _ in sorted_pages]
data = load_data('data.json')
user_id = 1
page_recommendation = find_pages_you_might_like(user_id,data)
print(page_recommendation)

            
        
```
