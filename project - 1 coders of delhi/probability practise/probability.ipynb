```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
```

```python
random = np.random.uniform(0,1,1000)
plt.hist(random , bins = 50 , density = True, color = 'blue',alpha = 0.6) 
plt.title('the uniform continuous data in between 0 to 1 ')
plt.xlabel('value')
plt.ylabel('probabibilty density')
plt.grid(True)
plt.legend(True)
plt.show()
```
