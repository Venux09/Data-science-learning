```python
import json 

def clean_data(data):
    #remove users with missing names  by using list comprehension 
     data["users"] =[ user for user in data["users"] if user["name"].strip()]#strip is used here because python counts blank spaces too /
    #ermove users with duplicate friends using list coprehension
     for user in data["users"]:
         
         user["friends"] = list(set(user["friends"]))
    
    #remove inactive users using lisy comprehension 
     data['users'] = [user for user in data['users'] if user['friends'] or user['liked_pages']]#strip is function which is used generally for the checking if there is gap taking the blank or not 
    # removing the duplicate pages 
     unique_id_pages = {}
     for page in data['pages']:#data of pages is page
         unique_id_pages[page['id']]  = page#setting the id of the data of page unique
     data['pages'] = list(unique_id_pages.values())#changing the set we have maded for unique id to list
     return data 
    

data = json.load(open('data2.json'))
data  = clean_data(data)
json.dump(data,open("cleaned_data.json","w") , indent = 4)
print("data cleaned suucessfully")
```
