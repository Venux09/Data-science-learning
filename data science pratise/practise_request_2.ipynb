```python
url = "https://httpbin.org/headers"
```

```python
headers= {
    'User-Agent':'Mozilla/5.0'
    'Accept-Language'
}


a = requests.get(url,headers=headers)
a.text
```

```python
#using the quesry params in the request module 
url = "https://httpbin.org/get"
```

```python
params = {
    'Search':'football',
    'page':2
}

response = requests.get(url,params=params)
print (response.url)
response.json()
```
