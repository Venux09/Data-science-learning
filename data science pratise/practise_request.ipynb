```python
import requests 
```

```python
response = requests.get('https://books.toscrape.com/')
```

```python
if response.status_code == 200:
    print(response.status_code)
    print('succesfully fetched the webiste')
else:
    print('failed to fetch the website')


with open(f"htmlss/pages.html",'w') as f :
    
    f.write(response.text)
```

```python
for i in range(1,51):
    
    a = requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html')
    with open(f"htmlss/pages--{i}.html",'w',encoding='utf-8') as f :
        
        f.write(a.text)
        print(f'Html file number {i} is downloaded in htmlss folder')
```
