```python
from bs4 import BeautifulSoup 
data=[]#list for the titles 
```

```python
for i in range(1,51):
    with open(f"Htmlss/page-{i}.html",encoding='utf-8') as f :
        content = f.read()
    #creating the Beautiful soup 
    soup = BeautifulSoup(content,"lxml")
    #product data 
    articles = soup.select("article.product_pod")
    #data scrapping
    for article in articles:
        title = article.find("h3").find("a")['title']#title of the books 
        price = article.select_one('p.price_color').text.split('£')[1]#price of the books 
        rating_element = article.select_one('p.star-rating')
        ratings = rating_element['class'][1]
        availability = article.select_one("p.instock.availability").text.strip()#using the strip for removing the extra new line line or gap 
        
        
        data.append([title,price,ratings,availability ])

```

```python
import pandas as pd 
df = pd.DataFrame(data,columns = ["Book Title","Price","Rating","Availability"])
df
```
