```python
import matplotlib.pyplot as plt 
```

```python
messi =   [60, 58, 52, 59, 54, 51, 50, 31, 38, 35]
ronaldo = [69, 61, 57, 55, 53, 49, 44, 37, 36, 34]
neymar =  [39, 43, 39, 49, 45, 37, 34, 19, 28, 24]
```

```python
plt.figure(figsize =(8,4))
plt.hist(messi,bins = 5,alpha= 1,label='messi',edgecolor='black')
plt.hist(ronaldo,bins=5,alpha = 1,label='ronaldo',edgecolor='blue')
plt.hist(neymar, bins = 5 , alpha = 1 , label = 'neymar',edgecolor='green')
plt.grid(True,linestyle='--',alpha=0.5)
plt.legend()
plt.title('data comparism of three footballer')

```
