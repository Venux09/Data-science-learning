```python
from sklearn import tree 
```

```python
#we are using the parameters in the way that it will judge the price weight and company name
```

```python
features = [[20,45,1],#company name - balaji
            [50,100,2],#company name - noise
            [25,100,3],#company name - lays"
            [100,500,4],#company name - bikaji
            [75,250,4]]#company name - Haldiram Mad angle 
```

```python
lables = ['Bhujia Namkeen','Nut cracker','Chips','Aloo bhujia','Nachos']
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
salesman = clf.fit(features,lables)
```

```python
print(f'''customer -: please give  this product according to the specifcation i have given you.
salesman -: ahh , okay please take the product according to the specificatoin you have given me { salesman.predict([[100,500,4]])}''')
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import plot_tree 
```

```python
plt.figure (figsize = (8,6))
plot_tree(salesman,feature_names = ["price","weight","company"],class_names = ["Bhujia Namkeen",
                                                                               "Nut cracker",
                                                                               "Chips",
                                                                              "Aloo bhujia",
                                                                              "Nachos"],filled = True)
plt.show()
```
