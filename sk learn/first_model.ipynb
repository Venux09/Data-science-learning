```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
feature = [[150,0], # 0 is for orange coulour and also for red cololur and also 1 is for the oragne coulolr 
           [170,0],
           [130,1],
           [120,1]]
```

```python
labels = ['apples', 'apples','oranges','oranges ']
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(feature,labels)
```

```python
from sklearn.tree import plot_tree #imorting plot tree to visulaize the decision tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
```

```python
plot_tree(clf,feature_names = ["weight","color"] , class_names = ["apples","oranges"], filled =True)
plt.show()
```
