```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')#used for the using styles in the charts bar or plotting line or the pie chart 
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0.2,0,0,0]
plt.title('Carrer runs of indiant batsman')
plt.pie(runs,labels=labels,colors = colors,wedgeprops = {'edgecolor':'black','linewidth':2,'linestyle':':'},shadow= True,explode = explode,startangle = 130,autopct ='%1.1f%%' )
plt.show()
```

```python
plt.pie(runs,colors = colors, labels = labels,explode = explode)
plt.title('Exploded pie chart of the data ')
plt.show()
```
