```python
import matplotlib.pyplot as plt
#data of the ronaldo carrier from 2013 to 2023
years = list(range(2013, 2023))
ronaldo = [69, 61, 57, 55, 53, 49, 44, 37, 36, 34]
```

```python
plt.bar(years,ronaldo)
plt.xlabel('Years - [2013-2023]')
plt.ylabel(' NUMBERS OF GOALS ')
plt.title('Performance of Cristano Ronaldo Year by Year')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
years = list(range(2013, 2023))
x = np.arange(len(years))#using a range of numpy here to make a fucntion for showing the data side by side in the comparisn chart 
width = 0.25

messi =   [60, 58, 52, 59, 54, 51, 50, 31, 38, 35]
ronaldo = [69, 61, 57, 55, 53, 49, 44, 37, 36, 34]
neymar =  [39, 43, 39, 49, 45, 37, 34, 19, 28, 24]
plt.bar(x-width,messi,width = width,label='MESSI')
plt.bar(x,ronaldo,width = width,label='Ronaldo')
plt.bar(x+width,neymar,width = width,label='Neymar')

plt.xlabel('Years')
plt.ylabel('GOAL SCORED')
plt.title('Comparison of the performance of the football players')
plt.xticks(x,years)#used for showing the actual years insted of the 1 , 2 ,3 
plt.legend()
plt.tight_layout()
plt.show()
```
