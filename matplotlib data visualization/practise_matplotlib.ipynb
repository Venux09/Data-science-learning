```python
import matplotlib.pyplot as plt 
```

```python
#data of football trio in their prime years 
```

```python
years = list(range(2013, 2023))

messi =   [60, 58, 52, 59, 54, 51, 50, 31, 38, 35]
ronaldo = [69, 61, 57, 55, 53, 49, 44, 37, 36, 34]
neymar =  [39, 43, 39, 49, 45, 37, 34, 19, 28, 24]
```

```python
#creating beautiful data visualization of the three goats
# plt.style('dark_background')
plt.plot(years,messi,'ro--',label = 'Prime messi goat')
plt.plot(years,ronaldo,'b.:',label = 'Prime ronaldo CR7 ')
plt.plot(years,neymar,'g*--',label = 'Prime neymar jr')
plt.xlabel('Years(2013-2023)')
plt.ylabel('--NO.OF GOALS-- ')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.title('Prime years of the CR7, messi and neymar jr')
plt.show()
```
