```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4,5,6]
y1 = [i*2 for i in x]
y2 =[i**2 for i in x]
y3 = [i**3 for i in x]
```

```python
#plotting the double of X data 
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('Double of X')
plt.xlabel('Number')
plt.ylabel('Double of the Number')
#plotting the square the X data using subplot
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Square of X')
plt.xlabel('Number')
plt.ylabel('square of the Number')
plt.tight_layout()
#cube of the X data using subplot for the comparism 
# plt.subplot(2,2,3)
# plt.plot(x,y3)
# plt.title('cube of X')
```

```python
# More variations of x
#2 is used here to define the  number of the rows 
#simple rule is defining the rowas and colums of the matrix and then defining the nummbrs of the graph
#here it is 1,2,3,4 and rows and columns is 2*2
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]
plt.figure(figsize=(8, 6)) # Optional: make it bigger
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')
plt.tight_layout()
plt.show()
```

```python
fig , axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(x,y1)
axs[0].set_title('x*2')
axs[1].plot(x,y2)
axs[1].set_title('x**2')
fig.suptitle('MY comaparism plot',fontsize=20)
fig.tight_layout()
fig.show()
```
