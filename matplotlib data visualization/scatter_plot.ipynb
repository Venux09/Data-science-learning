```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
colors = ['red' if score>50 else 'green' for score in exam_scores]
sizes = [score*2 for score in exam_scores]
```

```python
plt.scatter(study_hours,exam_scores,c = exam_scores,cmap = 'viridis')#c can be used as colors also and colors can also be used for the c 
plt.colorbar(label = 'score')#used for the getting the bar of  the color or cmap for the score
#colorbar can be changed for the different data also it is not related to the x and so it can used for the differnet data also like social behavioys 
plt.grid(True)
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1} ',(study_hours[i]+0.25,exam_scores[i]))
plt.xlabel("Hours of the studying")
plt.ylabel("NO. of the Marks")
plt.tight_layout()
plt.title('Data of the marks according to the hours the study')
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```
