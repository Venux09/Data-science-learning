```python
import matplotlib.pyplot as plt
```

```python
activities = ['playing','studying','sleeping',"Wacthing Tv"]
```

```python
time_spent = [2,3,4,5]
colors = ['blue','yellow','green','pink']
plt.figure(figsize = (6,6))
plt.pie(time_spent,labels = activities,colors = colors,autopct = '%1.1f%%',startangle = 90,shadow=True)
plt.axis('equal')
plt.title('Time spended by kids of the activities')
```

```python
days = [1, 2, 3, 4, 5, 6, 7] # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ["red", "blue", "green", "yellow"]

plt.figure(figsize = (6,6))
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels = labels ,colors = colors,alpha=0.4)
plt.legend(loc= 'upper left')#stackplot is used for knowing how much we are spending time on activites over the motth or over the week 
plt.title('Student time distribution')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.grid(True)
```
