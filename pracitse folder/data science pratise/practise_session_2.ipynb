```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data 
matches = [50, 52, 48, 55, 53, 49, 47, 45, 44, 42]

messi_goals =   [60, 58, 52, 59, 54, 51, 50, 31, 38, 35]
ronaldo_goals = [69, 61, 57, 55, 53, 49, 44, 37, 36, 34]
neymar_goals =  [39, 43, 39, 49, 45, 37, 34, 19, 28, 24]
```

```python
#going to show the scatter data 
```

```python
plt.scatter(matches,messi_goals,label='messi',cmap='viridis')
plt.scatter(matches,neymar_goals,label='neymar',cmap='viridis')
plt.scatter(matches,ronaldo_goals,label='ronaldo',cmap='viridis')
plt.colorbar(label='Goals')
plt.xlabel('Matches')
plt.ylabel('Goals')
plt.title('Matches palyed vs goals scored of footballers')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(matches, messi_goals, label='Messi', alpha=0.7)
plt.scatter(matches, neymar_goals, label='Neymar', alpha=0.7)
plt.scatter(matches, ronaldo_goals, label='Ronaldo', alpha=0.7)

# Trendline for Messi
z = np.polyfit(matches, messi_goals, 1)
p = np.poly1d(z)
plt.plot(matches, p(matches), linestyle='--')

plt.xlabel('Matches Played')
plt.ylabel('Goals Scored')
plt.title('Matches Played vs Goals Scored (Footballers)')

plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
