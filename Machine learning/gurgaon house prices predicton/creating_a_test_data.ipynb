```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('housing.csv')
```

```python
df
```

```python
df['income_cat'] = pd.cut(df['median_income'],
bins=[0,1.5,3,4.5,6,np.inf],
labels=[1,2,3,4,5])
```

```python
import matplotlib.pyplot as plt 
df['income_cat'].value_counts().sort_index().plot.bar(rot = 0, grid = True , color = 'black')
plt.style.use('seaborn-v0_8-whitegrid')
plt.title('Income Category Ditribution')
plt.xlabel('Income Category')
plt.ylabel('No of the instances ')#basically means no of objects that fall into that category 
plt.tight_layout()
```
