```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
```

```python
df['income_cat'] = pd.cut(df['median_income'],
bins=[0,1.5,3,4.5,6,np.inf],
labels=[1,2,3,4,5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit


split = StratifiedShuffleSplit(n_splits = 1 , test_size =  0.2,random_state  = 42 )


for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
# strat_train_set
```

```python
#removing the income_Cat column from the test and train data
```

```python
for sett in (strat_test_set,strat_train_set):
    sett.drop('income_cat',axis = 1,inplace = True )
```

```python
df = strat_train_set.copy()
```

```python
#one thing to remember is we never have to touch the data of the test set .
#we can only do our things with the data of the trainning data 
```

```python
import matplotlib.pyplot as plt 
df.plot(kind  = 'scatter',x = 'latitude',y='longitude',grid = True ,cmap = "jet" ,c = 'median_house_value')
plt.show 
```
