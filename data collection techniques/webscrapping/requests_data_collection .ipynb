```python
import requests
```

```python
a = requests.get('https://books.toscrape.com/')
```

```python
with open(f'htmls/1page.html','w') as  f :
    f.write(a.text)
```

```python
for i in range(1,51):
    a= requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html')
    with open(f'htmls/page{i}.html','w',encoding='utf-8') as f :
       f.write(a.text)
       print(f'page downloaded{i} successfully')
    
```
