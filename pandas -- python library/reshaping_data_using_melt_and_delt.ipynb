```python
import pandas as pd

# Sample DataFrame
data = {
'Name': ['Alice', 'Bob', 'Charlie'],
'Math': [85, 78, 92],
'Science': [90, 82, 89],
'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
#df melt is a method in which we can change or reshsape the entire data in differeent ways
dd = df.melt(id_vars = ['Name'] , value_vars = ['English','Math','Science'],var_name = 'Subjects',value_name = 'Score')
dd
```

```python
#pivot is a method is just uno reverse action of the melt function it brings back the data to wide format from the long format
```

```python
dd.pivot_table(index="Name", columns="Subjects",values="Score", aggfunc="mean")#have to use aggfunc in order to avoid the error because you cant have two same subject in this case with alice so yeah we have to use the agg func methode
```
