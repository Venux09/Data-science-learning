```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
 
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on='DeptID')#this is method used for merging the two data's
#it is called inner join method
```

```python
pd.merge(employees,departments,how="left")
```

```python
pd.merge(employees,departments, how="right")#this is the method which give every value of the right side and give nana in the missing places value 
```

```python
pd.merge(employees,departments,how="outer")#it is used for getting everyt output you can get out of the merged table 
#it give nan in the missing places 
```
