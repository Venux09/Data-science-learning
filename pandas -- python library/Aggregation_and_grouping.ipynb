```python
import pandas as pd
```

```python
df = pd.DataFrame({
"Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
"Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
"Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
"Salary": [85, 90, 78, 85, 92, 88, 75, 80],
"Age": [23, 25, 30, 22, 28, 26, 21, 27],
"JoinDate": pd.to_datetime([
"2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
"2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
])
})
```

```python
df
```

```python
#by using this groupby method it can be used in the to getting  the mean of the data.
(df.groupby('Team')['Salary'].mean())
```

```python
df.groupby('Team')['Salary'].min()#function for gettingthe output for the min 
```

```python
df.groupby('Team')['Salary'].max()#function for getting the max view
```

```python
df.groupby('Team')['Salary'].count()#useful for conting the number of people or objects in that rows
```

```python
df.groupby('Team')['Salary'].sum()#summing all the objects or values that are given in the rows
```

```python
df.groupby('Team')['Salary'].min()
```

```python
#way to get multiple out put in one time
df.groupby('Team')['Salary'].agg(['mean','max','min','count'])#agg and aggredate is same 
```

```python
df['Team average'] = df.groupby('Team')['Salary'].transform('mean')#transform is used to transform the data
df
```

```python
df['Team count'] = df.groupby('Team')['Salary'].transform('count')
```

```python
df
```

```python
#filter is the method which is used for the filtering the data such as giving the conditions 
df.groupby('Team').filter(lambda x: x['Salary'].mean()>85)
```
