```python
class choclates:
    company = "Amul"
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def choclates_detail(self):
        print(f"the choclate company owner is {self.name} and companyname is {self.company} his salalry is {self.salary}")


choco = choclates("cocomelon","4700000000")

choco.choclates_detail()
```

```python
class railway_form:
    train = "vande_bharat"
    def __init__(self,name,time,mobile_number,train_ticket,ticket_price):
        self.name = name
        self.time = time 
        self.mobile_number = mobile_number 
        self.train_ticket = train_ticket
        self.ticket_price = ticket_price
    def form_details(self):
        print(f"The name of the passanger is {self.name}\n The timing of the train when it  will deploy  is{self.time}\nthe mobile number of the passanger is {self.mobile_number}\nthe  no of the train ticket is {self.train_ticket}\nthe price of the ticket per person is {self.ticket_price}")
    @staticmethod
    def static_method():
        print(f"all details is done")
    @classmethod
    def class_method(cls):
        print(f"the name of the train is {cls.train}")
details = railway_form("monu verma","10:00","9994949##","1","2399")
details.form_details()
details.static_method()
details.class_method()              
```
