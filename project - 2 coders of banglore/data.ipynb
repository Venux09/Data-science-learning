```python
with open("final.txt", encoding = 'utf-8') as f :
    data = f.read()
```

```python
chunks = data.split('\n\n')#divides the data where there is some data with splits like \n\n
chunks = [c for c in chunks if len(c)>5]

```

```python
def parse_chunk(chunk):
    
        chunk = chunk.strip()
        sep_chunk = chunk.split('\n')
        username = sep_chunk[0]
        no_of_posts = int(sep_chunk[1].split(" posts")[0].replace(",","").replace(" post",""))
        no_of_followers = float(sep_chunk[2].split(" followers")[0].replace(",","").replace("K","").replace("m",""))
        if ("k" in sep_chunk[2]):
            no_of_followers = int(no_of_followers*1000)
        elif("M" in sep_chunk[2]):
            no_of_followers = int(no_of_followers*1000000)
            

        no_of_following = int(sep_chunk[3].split(" following")[0].replace(",",""))
        name = sep_chunk[4]
        if  (len(sep_chunk))>5:#for solving the problem the types of pages error and bio error for like every profile doesnot have bio and posts
            types_of_pages = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:])
        else:
            types_of_pages = "unknown"
            bio = ""
        return {"username":username,"no_of_followers":no_of_followers,"no_of_following":no_of_following,
                "no_of_posts":no_of_posts,"name":name,"types_of_pages":types_of_pages,"bio":bio}
```

```python
all_chunks = []
for chunk in chunks:
    
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
```

```python
import json 
s  = json.dumps(all_chunks,indent=4) 
with  open("data.json","w") as f :
    f.write(s)
```
